```python
numts_file = '../../NumtsGermline.txt'

with open(numts_file, 'r') as fp:
    data = fp.readlines()
Info = []
for lines in data:
    itemlist = lines.split(',')
    Info.append(itemlist)
Info1 = pd.DataFrame(Info)
Info1.columns = Info1.iloc[0,:]
Info1=Info1.iloc[1:, :]
Info1
```

```python
numts_file = '../../NumtsGermline.txt'

with open(numts_file, 'r') as fp:
    data = fp.readlines()

output_file = '../numts.germline.hg38.bed'
with open(output_file, 'w') as fp:
    for lines in data[1:]:
        itemlist = lines.split(',')
        fp.write(" ".join([itemlist[1][1:-1], itemlist[2], itemlist[3]]) + '\n')
```
